```python
data["Age"] = data["Age"].astype(float)

print(data.info())
```

```python
def convert_age_to_years(age_in_months):
    age_in_years = age_in_months / 12
    age_in_years = age_in_years.astype(int)
    return age_in_years
```

```python
data["Age in Years"] = convert_age_to_years(data["Age"])
```

```python
data.head()
```

```python
oldest_index = data["Age"].idxmax()
youngest_index = data["Age"].idxmin()

oldest_child_features = data.loc[oldest_index]

# Get the features of the youngest child
youngest_child_features = data.loc[youngest_index]

print("Features of the Oldest Child:")
print(oldest_child_features)

print("\nFeatures of the Youngest Child:")
print(youngest_child_features)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

age_column = data["Age"]

# Perform standardization
scaler = StandardScaler()
age_standardized = scaler.fit_transform(age_column.values.reshape(-1, 1))

# Perform normalization
normalizer = MinMaxScaler()
age_normalized = normalizer.fit_transform(age_column.values.reshape(-1, 1))

# Convert the scaled arrays back to a DataFrame
age_standardized = pd.DataFrame(age_standardized, columns=["Age (Standardized)"])
age_normalized = pd.DataFrame(age_normalized, columns=["Age (Normalized)"])

# Concatenate the scaled columns with the original data
scaled_data = pd.concat([data, age_standardized, age_normalized], axis=1)

# Display the scaled data
scaled_data.head()
```
