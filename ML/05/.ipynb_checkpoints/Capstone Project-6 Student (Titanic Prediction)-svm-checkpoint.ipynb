# SVC

- Support Vector Machine (SVM) is a supervised machine learning model for classifications and regressions

**SVM (Destek Vektör Makinesi)**, bir sınıflandırma ve regresyon algoritmasıdır. 

- Küçük veya orta büyüklükteki veri setleri için uygundur. 

- Distance base bir model olduğundan Scale’e duyarlıdır. Scale edilmesi gerekir.

- SVM'nin temel mantığı, verileri sınıflandırmak için verileri en iyi şekilde ayrıştıran bir hyperplane (hiperdüzlem) bulmaktır. 

- SVM, verileri bir uzayda temsil eder ve bu uzayda bir veya daha fazla hyperplane ile sınırları belirler. Hyperplane, verileri iki sınıfa ayıran en iyi ayrım çizgisidir. Bu çizgi, sınıflar arasındaki en büyük mesafeyi (margin) sağlamalıdır. SVM, bu margin mesafesini en büyük yapmak için çalışır.

- SVM ayrıca, doğrusal olmayan verilerle çalışırken kernel fonksiyonları kullanır. Kernel fonksiyonları, verilerin yüksek boyutlu uzaylarda temsil edilmesine izin verir. SVM, bu şekilde, daha karmaşık ve doğrusal olmayan verileri sınıflandırmak için daha yüksek doğruluk sağlar.

https://towardsdatascience.com/hyperparameter-tuning-for-support-vector-machines-c-and-gamma-parameters-6a5097416167
https://medium.com/grabngoinfo/support-vector-machine-svm-hyperparameter-tuning-in-python-a65586289bcb

**Peki Multiclass Olursa Ne Olur??**
- SVM algoritması, iki sınıflı sınıflandırma için tasarlanmıştır. Ancak, çok sınıflı sınıflandırma problemleri için de kullanılabilir. SVM, çok sınıflı sınıflandırma problemlerinde birden fazla iki sınıflı sınıflandırıcı kullanarak çalışır.

- Bu yaklaşım, "one-vs-all (rest)" (OvA) hepsine karşı bir veya "one-vs-one" (OvO) bire bir olarak bilinir. 

- OvA yaklaşımı, her sınıfın geri kalan tüm sınıflardan ayrı olarak ele alındığı ve sınıflandırıcının her sınıfı diğerlerinden ayırt etmeye çalıştığı bir yaklaşımdır. OvO yaklaşımı, her iki sınıf arasında bir sınıflandırıcı oluşturulur ve tüm sınıflandırıcıların çıktıları bir araya getirilerek en sık tekrarlanan sınıf belirlenir.

- Örneğin, 3 sınıflı bir sınıflandırma problemi için OvA yaklaşımı, önce sınıf 1'i diğerlerinden ayırmaya çalışan bir sınıflandırıcı oluşturur, ardından sınıf 2'yi diğerlerinden ayırmaya çalışan bir sınıflandırıcı oluşturur ve son olarak sınıf 3'ü diğerlerinden ayırmaya çalışan bir sınıflandırıcı oluşturur. Her sınıf için oluşturulan sınıflandırıcının çıktıları, hangi sınıfa ait olduğunu belirlemek için birleştirilir.

- OvO yaklaşımı ise her iki sınıf arasında bir sınıflandırıcı oluşturarak çalışır. Örneğin, 3 sınıflı bir sınıflandırma problemi için, sınıf 1 ve sınıf 2 arasında bir sınıflandırıcı, sınıf 1 ve sınıf 3 arasında bir sınıflandırıcı ve sınıf 2 ve sınıf 3 arasında bir sınıflandırıcı oluşturulur. Sonuç olarak, her bir çift sınıf arasındaki sınıflandırıcının çıktıları birleştirilir ve en sık tekrarlanan sınıf belirlenir.

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
svm_model_ilkel = SVC(random_state=101)
svm_model_ilkel.fit(X_train, y_train)
eval_metric(svm_model_ilkel, X_train, y_train, X_test, y_test)
```

**average parametresi**, çok sınıflı sınıflandırma problemlerinde bazı metriklerin (örneğin F1, precision, recall) nasıl hesaplanacağını belirtmek için kullanılır.

**average='macro':** Her sınıfın F1 puanlarının ağırlıklı olmayan ortalamasını döndürür. Bu yöntem, her sınıfın veri sayısının dengeli olması durumunda kullanılır.

**average='micro':** Tüm sınıfların yanıt sayılarının doğrudan toplanmasıyla, birleştirilmiş bir sınıf hesaplanır. Bu yöntem, tüm sınıfların veri sayısının dengesiz olduğu durumlarda kullanılır.

**average='weighted':** Her sınıfın F1 puanlarının, sınıfın örnek sayısıyla ağırlıklandırılmış ortalamasını döndürür. Yani her sınıfın ağırlığı, veri setindeki örnek sayısına göre belirlenir.

**average=None:** Her sınıf için bireysel skorları döndürür.

Özetle, macro yöntemi, her sınıfın performansını ayrı ayrı ölçer ve bu sınıfların performansını dengeli bir şekilde hesaplar. micro yöntemi ise, tüm sınıfların performansını birleştirir ve tüm sınıfların veri sayısının dengesiz olduğu durumlarda kullanılır.

accuracy_score fonksiyonunda average parametresi kullanılmaz çünkü doğruluk (accuracy) doğrudan tüm sınıflar üzerinden hesaplanır ve çok sınıflı sınıflandırmalarda sınıf dengesizliğine duyarlıdır.

## SVC GridsearchCV

**C:** Regülarizasyon parametresidir. Bu parametre ne kadar büyükse, sınıflandırma hatası o kadar düşük olur, ancak aynı zamanda overfitting riski de o kadar artar. Bu parametrenin küçük seçilmesi, daha iyi bir genelleştirme performansı sağlayabilir. C yi hard margin-soft margin belirlemek için kullanıyorduk. Model overfit olursa C’nin azaltılması gerekir.

**gamma:** Kernel parametresidir. RBF, polinom ve sigmoid kernellarında kullanılan kernel genişliğini kontrol eder. Data lineer değil de nonlineer ise datayı ayırmak içim kernel ekliyorduk. Düşük gamma değerleri daha geniş bir etki alanı tanımlar, bu nedenle karar sınırı daha düzgün olacaktır. Yüksek gamma değerleri daha kısıtlı bir etki alanı tanımlar, bu nedenle karar sınırı veriye daha yakın olabilir, ancak overfitting riski de artar.

Düşük bir gamma değeri, iki noktanın birbirine yakın olmasını tanımlar ve daha düzgün bir karar sınırı çizer. Yüksek bir gamma değeri ise, sadece çok yakın noktaların birbirine etkisi altında kalır ve daha dairesel bir karar sınırı çizer. gamma='scale' ile, veriler ölçeklendirilerek, farklı bir ölçekte olsalar bile, her bir özelliğin gamma için aynı ağırlığa sahip olması sağlanır

Model overfit olmuşsa gamma değerini düşürmemiz, model underfit olmuşsa gamma değerini yükseltmemiz gerekir.

**kernel:** Veri setinin doğrusal olarak ayrılamadığı durumlarda, veriyi daha yüksek boyutlu bir uzaya dönüştürerek doğrusal olarak ayrılabilir hale getirmek için kullanılır. Lineer, polinom, RBF (Gaussian) ve sigmoid gibi kernellar kullanılabilir.

**degree:** Sadece 'poly' yani polinom kernelı için kullanılır. Bu, polinomun derecesini belirtir. Örneğin, degree=2 bir ikinci dereceden polinomu temsil eder. Default değeri 3'tür.

```python
param_grid = {'SVC__C': [1, 1.3, 10],
              'SVC__gamma': ["scale", "auto", 0.2, 0.3],
              'SVC__kernel': ['rbf', 'linear']}

# C parametresi için ilk olarak denediğim değerler 'SVC__C': [0.001, 0.05, 0.01, 0.1]
```

```python
model = Pipeline([("scaler",scaler),("SVC", SVC(kernel="rbf", probability=True, class_weight="balanced", random_state=101))]) 

svm_model_grid = GridSearchCV(model,
                              param_grid,
                              verbose=2,
                              scoring="accuracy",
                              n_jobs=-1)

svm_model_grid.fit(X_train, y_train)

# default cv 5 olduğu için cv değeri yazmadım
# kernel="rbf" --> SVM'in doğrusal olarak ayrılamayan problemleri çözebilmesini sağlar.
```

## Optuna

Optuna, hiperparametre optimizasyonu için kullanılan açık kaynaklı bir Python kütüphanesidir. Optuna, veri bilimi projelerinde makine öğrenimi modelinin hiperparametrelerini otomatik olarak ayarlamak ve en iyi performansı elde etmek için kullanılır. Optuna'nın temel amacı, deneme ve hata yöntemlerine dayalı olarak hiperparametre uzayında etkili bir şekilde gezinmektir.

İşte Optuna'nın temel kavramları:

1- **Study (Çalışma):** Optuna'nın temel veri yapısıdır. Bir çalışma, belirli bir hiperparametre optimizasyon görevini temsil eder. Çalışma, hiperparametre değerlerinin belirli bir amaç fonksiyonunu (genellikle maksimize veya minimize edilen bir ölçüm metriği) nasıl etkilediğini analiz eder.

2- **Trial (Deneme):** Bir çalışmanın bir alt öğesidir. Bir deneme, belirli bir hiperparametre konfigürasyonunu temsil eder. Optuna, farklı denemelerde farklı hiperparametre değerlerini deneyerek en iyi sonuçları elde etmeye çalışır.

3- **Objective Function (Amaç Fonksiyonu):** Her deneme için optimize edilmek istenen ölçüm metriğini belirten fonksiyondur. Örneğin, doğruluk, ortalama mutlak hata veya farklı bir performans metriği olabilir.

4- **Sampler (Örnekleme Algoritması):** Optuna'nın hangi hiperparametre değerlerini deneyeceğini seçtiği algoritmadır. Örnekleme algoritmaları, hiperparametre uzayını gezme stratejilerini belirler.

Optuna'nın temel kullanım akışı şu şekildedir:

1- Bir çalışma oluşturun.
2- Her denemede belirli bir hiperparametre konfigürasyonu deneyin.
3- Amaç fonksiyonunu kullanarak denemeleri değerlendirin.
4- Optuna, daha iyi sonuç elde etmek için hiperparametre değerlerini güncelleyerek denemeleri yönlendirir.
5- Belirli bir durma koşuluna (örneğin, belirli sayıda deneme veya belirli bir süre) ulaşıldığında optimizasyonu sonlandırın.

**Optuna'nın avantajları**, hiperparametre optimizasyonunu otomatikleştirmesi, farklı örnekleme algoritmaları ve esnek yapılandırma seçenekleri sunmasıdır. Bu sayede veri bilimciler, hiperparametre ayarlaması konusunda daha etkili sonuçlar elde edebilirler.

```python
#pip install optuna
```

```python
import optuna
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
```

```python
def objective(trial):
    C = trial.suggest_float('C', 0.1, 10)
    kernel = trial.suggest_categorical('kernel', ['linear', 'poly', 'rbf', 'sigmoid'])
    gamma = trial.suggest_float('gamma', 0.0001, 1.0, log=True)
    
    model = SVC(
        C=C,
        kernel=kernel,
        gamma=gamma,
        random_state=42
    )
    
    return -cross_val_score(model, X_train, y_train, n_jobs=-1, cv=3, scoring='accuracy').mean()

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)
```

Bu kod, Optuna adlı bir hiperparametre optimizasyon kütüphanesi kullanarak bir SVC modelinin hiperparametrelerini en iyi performansı elde edecek şekilde ayarlamayı amaçlar. Aşağıda kodun adım adım açıklamasını bulabilirsiniz:

**1- def objective(trial) fonksiyonu:** Bu fonksiyon, hiperparametre optimizasyonu için bir amaç fonksiyonudur. Optuna'nın optimizasyon algoritması, bu fonksiyonu maksimize etmeye çalışır. Fonksiyonun içinde yer alan kod, her denemede farklı hiperparametre değerleri ile bir SVC modeli oluşturur ve çapraz doğrulama kullanarak modelin performansını ölçer. Fonksiyonun sonucu, çapraz doğrulama doğruluğunun negatifini döndürür. Bu, Optuna'nın en büyük değeri bulmaya çalışması gerektiği anlamına gelir.

**2- study = optuna.create_study(direction='maximize'):** Optuna'da bir çalışma (study) oluşturur. Bu çalışma, hiperparametre optimizasyonunun yapıldığı temel nesnedir. direction='maximize' parametresi, hedef amaç fonksiyonunun maksimize edilmeye çalışılacağını belirtir.

**3- study.optimize(objective, n_trials=100):** Bu satır, Optuna'nın hiperparametre optimizasyonunu gerçekleştirdiği ana kısımdır. objective fonksiyonunu, belirtilen sayıda deneme (n_trials) sayısı kadar çağırarak en iyi hiperparametre değerlerini bulmaya çalışır. Optuna'nın optimizasyon algoritmaları, farklı hiperparametre kombinasyonlarını deneyerek en yüksek amaç fonksiyonu değerini bulmaya çalışır.

Bu kod parçası, SVC modelinin hiperparametrelerini en iyi performansı sağlayacak şekilde otomatik olarak ayarlamak için Optuna'nın güçlü optimizasyon yeteneklerini kullanır. X_train ve y_train gibi değişkenleri, gerçek veriye ve hedef etiketlere karşılık gelen verilerle güncellemeyi unutmayın.

```python
best_params = study.best_params
best_model = SVC(
    C=best_params['C'],
    kernel=best_params['kernel'],
    gamma=best_params['gamma'],
    random_state=42
)
best_model.fit(X_train, y_train)
```

Bu kod, Optuna ile gerçekleştirilen hiperparametre optimizasyonunun sonuçlarına dayanarak en iyi hiperparametre değerleri ile bir SVC modelini oluşturmayı ve eğitmeyi amaçlar.

**1- best_params = study.best_params:** Bu satır, study nesnesinin best_params özelliğini kullanarak en iyi hiperparametre değerlerini alır. study nesnesi, optimize fonksiyonunun çağrıldığı sonunda en iyi hiperparametre değerlerini içerir.

**2- best_model = SVC(...):** Bu satır, en iyi hiperparametre değerlerini kullanarak bir SVC modeli oluşturur. best_params sözlüğünden C, kernel ve gamma değerlerini alarak SVC modelinin en iyi hiperparametre ayarlarıyla bir nesnesini oluşturur. Aynı zamanda random_state parametresi, modelin tekrarlanabilirliğini sağlamak için kullanılır.

**3- best_model.fit(X_train, y_train):** Oluşturulan best_modeli, eğitim verileri (X_train ve y_train) üzerinde eğitmek için kullanır. Model, verileri kullanarak sınıflandırma sınırlarını ve ayrımını öğrenir.

Bu kod parçası, Optuna tarafından bulunan en iyi hiperparametre değerlerine sahip bir SVC modelini oluşturur ve eğitir. Bu şekilde model, en iyi performansı elde etmek üzere otomatik olarak ayarlanmış olur. X_train ve y_train gibi değişkenleri, gerçek veriye ve hedef etiketlere karşılık gelen verilerle güncellemeyi unutmayın.

```python
y_pred_optuna_test = best_model.predict(X_test)
y_pred_optuna_train = best_model.predict(X_train)
```
