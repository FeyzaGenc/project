# SVC

- Support Vector Machine (SVM) is a supervised machine learning model for classifications and regressions

**SVM (Destek Vektr Makinesi)**, bir snflandrma ve regresyon algoritmasdr. 

- Kk veya orta byklkteki veri setleri iin uygundur. 

- Distance base bir model olduundan Scalee duyarldr. Scale edilmesi gerekir.

- SVM'nin temel mant, verileri snflandrmak iin verileri en iyi ekilde ayrtran bir hyperplane (hiperdzlem) bulmaktr. 

- SVM, verileri bir uzayda temsil eder ve bu uzayda bir veya daha fazla hyperplane ile snrlar belirler. Hyperplane, verileri iki snfa ayran en iyi ayrm izgisidir. Bu izgi, snflar arasndaki en byk mesafeyi (margin) salamaldr. SVM, bu margin mesafesini en byk yapmak iin alr.

- SVM ayrca, dorusal olmayan verilerle alrken kernel fonksiyonlar kullanr. Kernel fonksiyonlar, verilerin yksek boyutlu uzaylarda temsil edilmesine izin verir. SVM, bu ekilde, daha karmak ve dorusal olmayan verileri snflandrmak iin daha yksek doruluk salar.

- SVM algoritmas, iki snfl snflandrma iin tasarlanmtr. Ancak, ok snfl snflandrma problemleri iin de kullanlabilir. SVM, ok snfl snflandrma problemlerinde birden fazla iki snfl snflandrc kullanarak alr.

- Bu yaklam, "one-vs-all (rest)" (OvA) hepsine kar bir veya "one-vs-one" (OvO) bire bir olarak bilinir. 

- OvA yaklam, her snfn geri kalan tm snflardan ayr olarak ele alnd ve snflandrcnn her snf dierlerinden ayrt etmeye alt bir yaklamdr. OvO yaklam, her iki snf arasnda bir snflandrc oluturulur ve tm snflandrclarn ktlar bir araya getirilerek en sk tekrarlanan snf belirlenir.

- rnein, 3 snfl bir snflandrma problemi iin OvA yaklam, nce snf 1'i dierlerinden ayrmaya alan bir snflandrc oluturur, ardndan snf 2'yi dierlerinden ayrmaya alan bir snflandrc oluturur ve son olarak snf 3' dierlerinden ayrmaya alan bir snflandrc oluturur. Her snf iin oluturulan snflandrcnn ktlar, hangi snfa ait olduunu belirlemek iin birletirilir.

- OvO yaklam ise her iki snf arasnda bir snflandrc oluturarak alr. rnein, 3 snfl bir snflandrma problemi iin, snf 1 ve snf 2 arasnda bir snflandrc, snf 1 ve snf 3 arasnda bir snflandrc ve snf 2 ve snf 3 arasnda bir snflandrc oluturulur. Sonu olarak, her bir ift snf arasndaki snflandrcnn ktlar birletirilir ve en sk tekrarlanan snf belirlenir.

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
svm_model_ilkel = SVC(random_state=101)
svm_model_ilkel.fit(X_train, y_train)
eval_metric(svm_model_ilkel, X_train, y_train, X_test, y_test)
```

**average parametresi**, ok snfl snflandrma problemlerinde baz metriklerin (rnein F1, precision, recall) nasl hesaplanacan belirtmek iin kullanlr.

**average='macro':** Her snfn F1 puanlarnn arlkl olmayan ortalamasn dndrr. Bu yntem, her snfn veri saysnn dengeli olmas durumunda kullanlr.

**average='micro':** Tm snflarn yant saylarnn dorudan toplanmasyla, birletirilmi bir snf hesaplanr. Bu yntem, tm snflarn veri saysnn dengesiz olduu durumlarda kullanlr.

**average='weighted':** Her snfn F1 puanlarnn, snfn rnek saysyla arlklandrlm ortalamasn dndrr. Yani her snfn arl, veri setindeki rnek saysna gre belirlenir.

**average=None:** Her snf iin bireysel skorlar dndrr.

zetle, macro yntemi, her snfn performansn ayr ayr ler ve bu snflarn performansn dengeli bir ekilde hesaplar. micro yntemi ise, tm snflarn performansn birletirir ve tm snflarn veri saysnn dengesiz olduu durumlarda kullanlr.

accuracy_score fonksiyonunda average parametresi kullanlmaz nk doruluk (accuracy) dorudan tm snflar zerinden hesaplanr ve ok snfl snflandrmalarda snf dengesizliine duyarldr.

## SVC GridsearchCV

**C:** Reglarizasyon parametresidir. Bu parametre ne kadar bykse, snflandrma hatas o kadar dk olur, ancak ayn zamanda overfitting riski de o kadar artar. Bu parametrenin kk seilmesi, daha iyi bir genelletirme performans salayabilir. C yi hard margin-soft margin belirlemek iin kullanyorduk. Model overfit olursa Cnin azaltlmas gerekir.

**gamma:** Kernel parametresidir. RBF, polinom ve sigmoid kernellarnda kullanlan kernel geniliini kontrol eder. Data lineer deil de nonlineer ise datay ayrmak iim kernel ekliyorduk. Dk gamma deerleri daha geni bir etki alan tanmlar, bu nedenle karar snr daha dzgn olacaktr. Yksek gamma deerleri daha kstl bir etki alan tanmlar, bu nedenle karar snr veriye daha yakn olabilir, ancak overfitting riski de artar.

Dk bir gamma deeri, iki noktann birbirine yakn olmasn tanmlar ve daha dzgn bir karar snr izer. Yksek bir gamma deeri ise, sadece ok yakn noktalarn birbirine etkisi altnda kalr ve daha dairesel bir karar snr izer. gamma='scale' ile, veriler leklendirilerek, farkl bir lekte olsalar bile, her bir zelliin gamma iin ayn arla sahip olmas salanr

Model overfit olmusa gamma deerini drmemiz, model underfit olmusa gamma deerini ykseltmemiz gerekir.

**kernel:** Veri setinin dorusal olarak ayrlamad durumlarda, veriyi daha yksek boyutlu bir uzaya dntrerek dorusal olarak ayrlabilir hale getirmek iin kullanlr. Lineer, polinom, RBF (Gaussian) ve sigmoid gibi kernellar kullanlabilir.

**degree:** Sadece 'poly' yani polinom kernel iin kullanlr. Bu, polinomun derecesini belirtir. rnein, degree=2 bir ikinci dereceden polinomu temsil eder. Default deeri 3'tr.

```python
param_grid = {'SVC__C': [1, 1.3, 10],
              'SVC__gamma': ["scale", "auto", 0.2, 0.3],
              'SVC__kernel': ['rbf', 'linear']}

# C parametresi iin ilk olarak denediim deerler 'SVC__C': [0.001, 0.05, 0.01, 0.1]
```

```python
model = Pipeline([("scaler",scaler),("SVC", SVC(kernel="rbf", probability=True, class_weight="balanced", random_state=101))]) 

svm_model_grid = GridSearchCV(model,
                              param_grid,
                              verbose=2,
                              scoring="accuracy",
                              n_jobs=-1)

svm_model_grid.fit(X_train, y_train)

# default cv 5 olduu iin cv deeri yazmadm
# kernel="rbf" --> SVM'in dorusal olarak ayrlamayan problemleri zebilmesini salar.
```

## Optuna

Optuna, hiperparametre optimizasyonu iin kullanlan ak kaynakl bir Python ktphanesidir. Optuna, veri bilimi projelerinde makine renimi modelinin hiperparametrelerini otomatik olarak ayarlamak ve en iyi performans elde etmek iin kullanlr. Optuna'nn temel amac, deneme ve hata yntemlerine dayal olarak hiperparametre uzaynda etkili bir ekilde gezinmektir.

te Optuna'nn temel kavramlar:

1- **Study (alma):** Optuna'nn temel veri yapsdr. Bir alma, belirli bir hiperparametre optimizasyon grevini temsil eder. alma, hiperparametre deerlerinin belirli bir ama fonksiyonunu (genellikle maksimize veya minimize edilen bir lm metrii) nasl etkilediini analiz eder.

2- **Trial (Deneme):** Bir almann bir alt esidir. Bir deneme, belirli bir hiperparametre konfigrasyonunu temsil eder. Optuna, farkl denemelerde farkl hiperparametre deerlerini deneyerek en iyi sonular elde etmeye alr.

3- **Objective Function (Ama Fonksiyonu):** Her deneme iin optimize edilmek istenen lm metriini belirten fonksiyondur. rnein, doruluk, ortalama mutlak hata veya farkl bir performans metrii olabilir.

4- **Sampler (rnekleme Algoritmas):** Optuna'nn hangi hiperparametre deerlerini deneyeceini setii algoritmadr. rnekleme algoritmalar, hiperparametre uzayn gezme stratejilerini belirler.

Optuna'nn temel kullanm ak u ekildedir:

1- Bir alma oluturun.
2- Her denemede belirli bir hiperparametre konfigrasyonu deneyin.
3- Ama fonksiyonunu kullanarak denemeleri deerlendirin.
4- Optuna, daha iyi sonu elde etmek iin hiperparametre deerlerini gncelleyerek denemeleri ynlendirir.
5- Belirli bir durma kouluna (rnein, belirli sayda deneme veya belirli bir sre) ulaldnda optimizasyonu sonlandrn.

**Optuna'nn avantajlar**, hiperparametre optimizasyonunu otomatikletirmesi, farkl rnekleme algoritmalar ve esnek yaplandrma seenekleri sunmasdr. Bu sayede veri bilimciler, hiperparametre ayarlamas konusunda daha etkili sonular elde edebilirler.

```python
#pip install optuna
```

```python
import optuna
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
```

```python
def objective(trial):
    C = trial.suggest_float('C', 0.1, 10)
    kernel = trial.suggest_categorical('kernel', ['linear', 'poly', 'rbf', 'sigmoid'])
    gamma = trial.suggest_float('gamma', 0.0001, 1.0, log=True)
    
    model = SVC(
        C=C,
        kernel=kernel,
        gamma=gamma,
        random_state=42
    )
    
    return -cross_val_score(model, X_train, y_train, n_jobs=-1, cv=3, scoring='accuracy').mean()

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)
```

Bu kod, Optuna adl bir hiperparametre optimizasyon ktphanesi kullanarak bir SVC modelinin hiperparametrelerini en iyi performans elde edecek ekilde ayarlamay amalar. Aada kodun adm adm aklamasn bulabilirsiniz:

**1- def objective(trial) fonksiyonu:** Bu fonksiyon, hiperparametre optimizasyonu iin bir ama fonksiyonudur. Optuna'nn optimizasyon algoritmas, bu fonksiyonu maksimize etmeye alr. Fonksiyonun iinde yer alan kod, her denemede farkl hiperparametre deerleri ile bir SVC modeli oluturur ve apraz dorulama kullanarak modelin performansn ler. Fonksiyonun sonucu, apraz dorulama doruluunun negatifini dndrr. Bu, Optuna'nn en byk deeri bulmaya almas gerektii anlamna gelir.

**2- study = optuna.create_study(direction='maximize'):** Optuna'da bir alma (study) oluturur. Bu alma, hiperparametre optimizasyonunun yapld temel nesnedir. direction='maximize' parametresi, hedef ama fonksiyonunun maksimize edilmeye allacan belirtir.

**3- study.optimize(objective, n_trials=100):** Bu satr, Optuna'nn hiperparametre optimizasyonunu gerekletirdii ana ksmdr. objective fonksiyonunu, belirtilen sayda deneme (n_trials) says kadar ararak en iyi hiperparametre deerlerini bulmaya alr. Optuna'nn optimizasyon algoritmalar, farkl hiperparametre kombinasyonlarn deneyerek en yksek ama fonksiyonu deerini bulmaya alr.

Bu kod paras, SVC modelinin hiperparametrelerini en iyi performans salayacak ekilde otomatik olarak ayarlamak iin Optuna'nn gl optimizasyon yeteneklerini kullanr. X_train ve y_train gibi deikenleri, gerek veriye ve hedef etiketlere karlk gelen verilerle gncellemeyi unutmayn.

```python
best_params = study.best_params
best_model = SVC(
    C=best_params['C'],
    kernel=best_params['kernel'],
    gamma=best_params['gamma'],
    random_state=42
)
best_model.fit(X_train, y_train)
```

Bu kod, Optuna ile gerekletirilen hiperparametre optimizasyonunun sonularna dayanarak en iyi hiperparametre deerleri ile bir SVC modelini oluturmay ve eitmeyi amalar.

**1- best_params = study.best_params:** Bu satr, study nesnesinin best_params zelliini kullanarak en iyi hiperparametre deerlerini alr. study nesnesi, optimize fonksiyonunun arld sonunda en iyi hiperparametre deerlerini ierir.

**2- best_model = SVC(...):** Bu satr, en iyi hiperparametre deerlerini kullanarak bir SVC modeli oluturur. best_params szlnden C, kernel ve gamma deerlerini alarak SVC modelinin en iyi hiperparametre ayarlaryla bir nesnesini oluturur. Ayn zamanda random_state parametresi, modelin tekrarlanabilirliini salamak iin kullanlr.

**3- best_model.fit(X_train, y_train):** Oluturulan best_modeli, eitim verileri (X_train ve y_train) zerinde eitmek iin kullanr. Model, verileri kullanarak snflandrma snrlarn ve ayrmn renir.

Bu kod paras, Optuna tarafndan bulunan en iyi hiperparametre deerlerine sahip bir SVC modelini oluturur ve eitir. Bu ekilde model, en iyi performans elde etmek zere otomatik olarak ayarlanm olur. X_train ve y_train gibi deikenleri, gerek veriye ve hedef etiketlere karlk gelen verilerle gncellemeyi unutmayn.

```python
y_pred_optuna_test = best_model.predict(X_test)
y_pred_optuna_train = best_model.predict(X_train)
```
